```
class Nodo:
    def __init__(self ,value , siguiente = None):
        self.data = value       
        self.siguiente = siguiente

class ListaCircular:
    def __init__(self):  
        self.__head = None 
        self.__tail = None 
        self.__size = 0 

    def is_empty (self):
        return self.__size == 0

    def get_size(self):
        return self.__size

    def insert (self, value):
        if self.is_empty() :
            self.__tail = self.__head= Nodo(value)
            self.__head.siguiente= self.__tail
            self.__size +=1

        elif self.search(value) == True :
              print("Error")
            
        elif value > self.__head.data:
            new_node = self.__head
            self.__head = new_node.siguiente= Nodo(value, self.__tail)
            self.__size +=1

        elif value < self.__head.data and value < self.__tail.data:
             new_node =  Nodo(value)
             new_node.siguiente = self.__tail
             self.__tail = new_node 
             self.__head.siguiente = new_node
             self.__size +=1

        elif value < self.__head.data and  value > self.__tail.data:
            new_node = Nodo(value)
            curr_node= self.__tail
            while new_node.data > curr_node.siguiente.data:
                curr_node = curr_node.siguiente
            new_node.siguiente= curr_node.siguiente
            curr_node.siguiente= new_node 
            self.__size +=1

    def search (self, value):
         new_node = Nodo(value)
         curr_node= self.__head
         while new_node.data != curr_node.siguiente.data and curr_node.siguiente != self.__head:  
                curr_node = curr_node.siguiente
         if new_node.data == curr_node.siguiente.data:
             return True
         else: 
             return False

    def remove(self, value): 
        new_node = Nodo(value)
        curr_node= self.__head
        if self.search(value) == False:
              print("Imposible eliminar")
        else:
            while new_node.data != curr_node.siguiente.data and curr_node.siguiente != self.__head:  
                  curr_node = curr_node.siguiente
            if curr_node.siguiente.data == self.__head.data:
                curr_node.siguiente =  self.__head.siguiente
                self.__head = curr_node
                self.__size -=1
               
            elif curr_node.siguiente.data == self.__tail.data:
                self.__tail = self.__tail.siguiente
                self.__head.siguiente = self.__tail
                self.__size -=1
                     
            elif curr_node.siguiente.data != self.__head.data and curr_node.siguiente.data != self.__tail.data:
                anterior = None 
                while curr_node.data != value and curr_node.siguiente != self.__head:
                      anterior = curr_node
                      curr_node= curr_node.siguiente
                if curr_node.data == value:
                    anterior.siguiente= curr_node.siguiente
                    self.__size -=1

    def tranversal (self):
        curr_node = self.__head
        while curr_node.siguiente != self.__head:
            print(f"{ curr_node.siguiente.data } --> " , end="")
            curr_node= curr_node.siguiente
        print(self.__head.data)
        print("")
       



#Pruebas
l = ListaCircular()
print(f"La lista est치 vacia? {l.is_empty()}")
l.insert(3)
l.insert(6)
l.insert(2)
l.insert(56)
l.insert(14)
l.insert(1)
l.insert(28)
l.insert(194)
l.insert(34)
l.insert(47)
print(f"La lista est치 vacia? {l.is_empty()}")
print(f"La lista se compone de {l.get_size()} elementos")
l.tranversal()
l.remove(6)
l.remove(28)
l.remove(1)
l.tranversal()
print(f"La lista se compone de {l.get_size()} elementos")
l.search(3)
l.insert(14)
```
